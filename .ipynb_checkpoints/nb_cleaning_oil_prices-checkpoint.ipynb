```python
import numpy as np
import pandas as pd
import re

import matplotlib.pyplot as plt
import statsmodels.api as sm

from scipy.interpolate import interp1d
```

```python
gas_prices = pd.read_csv("./datasets/gasoline-prices-by-country.csv")
chad_gas_prices = gas_prices[gas_prices["Country Name"] == "Chad"]
chad_gas_prices
```

```python
plt.figure(figsize=(14,7))
plt.scatter(chad_gas_prices['year'], chad_gas_prices['gas prices'], s=100, marker="o", c="b")
plt.scatter(chad_gas_prices['year'], chad_gas_prices['filtered gas prices'], s=75, marker=".", c="r")
#plt.plot(chad_gas_prices['year'], chad_gas_prices["gas prices"], 'bo')
#plt.plot(chad_gas_prices['year'], chad_gas_prices["filtered gas prices"], 'r*')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Gas price (US$ per liter)", fontsize=12)
plt.legend(["original data", "filtered data"], loc="upper right")
plt.title("Chad gas price per year", fontsize=20)
plt.savefig("gas-price-filter-vs-nofilter.png")
plt.show()
```
