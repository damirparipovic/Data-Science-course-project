```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats
```

```python
barrel_to_liter = (1/159)
```

```python
# take the crude oil price data and convert it to the US$ per liter for the years from 1995 - 2016
crude_df = pd.read_csv("./datasets/crude-oil-prices.csv")
crude_df = crude_df.drop(["Code", "Entity"], axis=1).rename(columns={"Oil - Crude prices since 1861 (current $)":"Crude prices US$ per barrel"})
crude_df = crude_df.rename(columns={"Year":"year"})
crude_df = crude_df[(crude_df["year"] >= 1995) & (crude_df["year"] <= 2016)]
crude_df["Crude US$ per liter"] = crude_df["Crude prices US$ per barrel"] * barrel_to_liter
crude_df = crude_df.set_index("year")
```

```python
# load the previously generated gasoline prices data set
gas_df = pd.read_csv("./datasets/gasoline-prices-by-country.csv")
gas_df = gas_df.drop(["gas prices"], axis=1)
#gas_df
```

```python
# Get the countries that have valid values for the filtered gas prices, they have data from 1995 to 2016
country_check_year = gas_df[gas_df["year"] == 1995]
valid_countries = country_check_year[country_check_year["filtered gas prices"].notna()]
countries_list = valid_countries["Country Name"].values.tolist()
```

```python
# Collect all the data for the countries found in the cell above
gas_valid_country = gas_df[gas_df["Country Name"].isin(countries_list)]
gas_valid_country = gas_valid_country.set_index("Country Name")
gas_valid_country = gas_valid_country[gas_valid_country["year"] >= 1995]
#gas_valid_country
```

```python
# merge the two dataframes
gas_valid_country["crude price US$"] = gas_valid_country.apply(lambda row: crude_df["Crude US$ per liter"].loc[row["year"]], axis=1)
```

```python
# find the difference between the gas price (at pump) and the crude oil price
gas_valid_country["diff in price US$"] = gas_valid_country["filtered gas prices"] - gas_valid_country["crude price US$"]
gas_valid_country["diff in price US$ (y + 1)"] = gas_valid_country["diff in price US$"].shift(1)
gas_valid_country.loc[gas_valid_country['year'] == 1995, 'diff in price US$ (y + 1)'] = np.nan
gas_valid_country["year to year diff in price"] = gas_valid_country["diff in price US$"] - gas_valid_country["diff in price US$ (y + 1)"]
```

```python
# group 0
germany = gas_valid_country.loc["Germany"]
denmark = gas_valid_country.loc["Denmark"]
israel = gas_valid_country.loc["Israel"]

# group 1
us = gas_valid_country.loc["United States"]
canada = gas_valid_country.loc["Canada"]
china = gas_valid_country.loc["China"]

# group 2
kenya = gas_valid_country.loc["Kenya"]
serbia = gas_valid_country.loc["Serbia"]
lithuania = gas_valid_country.loc["Lithuania"]


plt.figure(figsize=(14,7))
plt.plot(us['year'], us["filtered gas prices"], 'bo')
plt.plot(germany['year'], germany["filtered gas prices"], 'yo')
plt.plot(kenya['year'], kenya["filtered gas prices"], 'ro')
plt.plot(us['year'], us["crude price US$"], 'k-')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Gas price (US$ per liter)", fontsize=12)
plt.legend(["United States", "Germany",  "Kenya", "Crude price"], loc="upper left")
plt.title("Gas price per year for different countries", fontsize=20)
plt.savefig("gas-price-for-three-countries.png", dpi=180)
plt.show()
```

```python
plt.figure(figsize=(20,8)).suptitle("Showing gas prices per year from 3 different countries for each group", fontsize=18)
plt.subplot(1,3,1)
plt.xlabel("Year", fontsize=13)
plt.ylabel("Gas price(US$/L)", fontsize=13)
plt.ylim(top=2.2)
plt.title("High price group", fontsize=15)
plt.plot(germany['year'], germany["filtered gas prices"], 'b-')
plt.plot(denmark['year'], denmark["filtered gas prices"], 'r-')
plt.plot(israel['year'], israel["filtered gas prices"], 'y-')
plt.plot(us['year'], us["crude price US$"], 'k-')
plt.legend(["Germany", "Denmark",  "Israel", "Crude price"], loc="upper left")
#plt.legend(["United States", "Canada", "Germany", "Crude price"], loc="upper left")
plt.subplot(1,3,2)
plt.xlabel("Year", fontsize=13)
plt.ylabel("Gas price(US$/L)", fontsize=13)
plt.ylim(top=2.2)
plt.title("Low price group", fontsize=15)
plt.plot(us['year'], us["filtered gas prices"], 'b-')
plt.plot(canada['year'], canada["filtered gas prices"], 'r-')
plt.plot(china['year'], china["filtered gas prices"], 'y-')
plt.plot(us['year'], us["crude price US$"], 'k-')
plt.legend(["United States", "Canada",  "China", "Crude price"], loc="upper left")

plt.subplot(1,3,3)
plt.xlabel("Year", fontsize=13)
plt.ylabel("Gas price(US$/L)", fontsize=13)
plt.ylim(top=2.2)
plt.title("Mid price group", fontsize=15)
plt.plot(lithuania['year'], lithuania["filtered gas prices"], 'b-')
plt.plot(kenya['year'], kenya["filtered gas prices"], 'r-')
plt.plot(serbia['year'], serbia["filtered gas prices"], 'y-')
plt.plot(us['year'], us["crude price US$"], 'k-')
plt.legend(["Lithuania", "Kenya",  "Serbia", "Crude price"], loc="upper left")

plt.savefig("3-country-reps-per-group.png", dpi=180)
plt.show()
```

```python
# tax inspection USA 2012 - 2016
us_5_yr = us[us["year"] >= 2012]
us_5_yr["implied tax + corporate fees"] = us_5_yr["diff in price US$"] / us_5_yr["filtered gas prices"] * 100
lin_reg_us_5_yr = stats.linregress(us_5_yr["year"], us_5_yr["implied tax + corporate fees"])
print("The slope of US difference in price " + str(lin_reg_us_5_yr.slope))
print("The R-value of US " + str(lin_reg_us_5_yr.rvalue))
```
